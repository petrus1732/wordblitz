```python
import json
from datetime import datetime

ISO_FORMAT = '%Y-%m-%d'

def parse_date(value):
    if not value:
        return None
    try:
        return datetime.strptime(value, ISO_FORMAT)
    except ValueError:
        return None

def normalize_name(value):
    return (value or '').strip().lower()

with open('event_rankings.json', 'r', encoding='utf-8') as fh:
    events = json.load(fh)

indexed_events = list(enumerate(events))
indexed_events.sort(key=lambda entry: (parse_date(entry[1].get('date')) or datetime.max, entry[0]))

filtered_events = []
last_kept_date = {}
for original_index, event in indexed_events:
    event_name = normalize_name(event.get('name'))
    event_date = parse_date(event.get('date'))
    previous_date = last_kept_date.get(event_name)
    if event_date and previous_date and (event_date - previous_date).days <= 1:
        # duplicate caused by timezone; keep the earlier copy that already exists
        continue
    filtered_events.append(event)
    if event_date:
        last_kept_date[event_name] = event_date

def clean_rankings(rankings):
    deduped = []
    seen_ids = set()
    for entry in rankings or []:
        key = entry.get('playerId') or f"name:{entry.get('name', 'Unknown')}"
        if key in seen_ids:
            continue
        seen_ids.add(key)
        deduped.append(dict(entry))

    arenas = []
    others = []
    for entry in deduped:
        if normalize_name(entry.get('name')) == 'all arenas':
            entry['rank'] = 0
            arenas.append(entry)
        else:
            others.append(entry)

    others.sort(key=lambda item: item.get('points') or 0, reverse=True)
    for idx, entry in enumerate(others, start=1):
        entry['rank'] = idx
    return others + arenas

cleaned_events = []
for event in filtered_events:
    cleaned_events.append({**event, 'rankings': clean_rankings(event.get('rankings'))})
cleaned_events.sort(key=lambda event: parse_date(event.get('date')), reverse=True)
with open('event_rankings.json', 'w', encoding='utf-8') as fh:
    json.dump(cleaned_events, fh, ensure_ascii=False, indent=2)

len(cleaned_events)
for event in cleaned_events:
    print(event['date'], event['name'])
    for person in event['rankings']:
        print(f"  {person['rank']}. {person['name']} - {person['points']}")
```

```python

import csv
import json
from collections import defaultdict


def dense_rank_entries(entries):
    arenas = []
    others = []
    for entry in entries:
        name = (entry.get('name') or '').strip()
        if name.lower() == 'all arenas':
            entry['rank'] = 0
            arenas.append(entry)
        else:
            others.append(entry)
    others.sort(key=lambda item: (-(item.get('points') or 0), item.get('name') or ''))
    current_rank = 0
    previous_points = None
    for entry in others:
        points = entry.get('points') or 0
        if previous_points is None or points != previous_points:
            current_rank += 1
            previous_points = points
        entry['rank'] = current_rank
    return others + arenas


with open('event_rankings.json', 'r', encoding='utf-8') as fh:
    event_data = json.load(fh)
for event in event_data:
    rankings = event.get('rankings') or []
    # ensure numeric points
    for row in rankings:
        if isinstance(row.get('points'), str):
            row['points'] = int(row['points'].replace(',', '') or 0)
    event['rankings'] = dense_rank_entries(rankings)
with open('event_rankings.json', 'w', encoding='utf-8') as fh:
    json.dump(event_data, fh, ensure_ascii=False, indent=2)


def load_daily_rows():
    with open('daily_scores.csv', newline='', encoding='utf-8') as fh:
        reader = csv.DictReader(fh)
        return list(reader)


def save_daily_rows(rows):
    fieldnames = ['dailyDate', 'rank', 'playerId', 'name', 'points', 'avatarUrl']
    with open('daily_scores.csv', 'w', newline='', encoding='utf-8') as fh:
        writer = csv.DictWriter(fh, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(rows)


def dense_rank_daily(rows):
    grouped = defaultdict(list)
    for row in rows:
        row['points'] = int(str(row.get('points') or '0').replace(',', '') or 0)
        grouped[row['dailyDate']].append(row)

    cleaned = []
    for date in sorted(grouped.keys()):
        entries = grouped[date]
        entries.sort(key=lambda item: (-item['points'], item.get('name') or ''))
        current_rank = 0
        previous_points = None
        arenas = []
        for entry in entries:
            if (entry.get('name') or '').strip().lower() == 'all arenas':
                entry['rank'] = ''
                arenas.append(entry)
                continue
            if previous_points is None or entry['points'] != previous_points:
                current_rank += 1
                previous_points = entry['points']
            entry['rank'] = str(current_rank)
            cleaned.append(entry)
        for arena in arenas:
            cleaned.append(arena)
    return cleaned


daily_rows = load_daily_rows()
updated_daily = dense_rank_daily(daily_rows)
# convert points back to string for CSV consistency
for row in updated_daily:
    row['points'] = str(row['points'])
save_daily_rows(updated_daily)

print(f"Updated {len(event_data)} events and {len(updated_daily)} daily rows with dense ranks.")

```
